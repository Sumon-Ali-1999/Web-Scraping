```python
import pandas as pd
df=pd.DataFrame()
ljob_title=[]
lexperience_level=[]
ljob_skill=[]
```

```python
jobs_block=driver.find_elements(By.XPATH,"//ul[@class='scaffold-layout__list-container']/li")

#scrool view
for job in jobs_block:
    driver.execute_script("arguments[0].scrollIntoView()",job)
    time.sleep(.5)

job_links=driver.find_elements(By.XPATH,"//ul[@class='scaffold-layout__list-container']/li/div/div/div[1]/div[2]/div[1]/a")
for link in job_links:
    link.click()
    time.sleep(3)
    job_title=driver.find_element(By.XPATH,"//h2[@class='t-24 t-bold job-details-jobs-unified-top-card__job-title']").text
    experience_level=driver.find_element(By.XPATH,"(//span[@class='job-details-jobs-unified-top-card__job-insight-view-model-secondary'])[1]").text
    job_skill=driver.find_element(By.XPATH,"//button[@class='job-details-jobs-unified-top-card__job-insight-text-button']").text
    
    ljob_title.append(job_title)
    lexperience_level.append(experience_level)
    ljob_skill.append(job_skill)
```

```python
data={'job_title':ljob_title,'experience_level':lexperience_level,'job_skill':ljob_skill}
df=pd.DataFrame(data)
df
```

```python
df.to_csv('scrapping.csv',index=False)
```
